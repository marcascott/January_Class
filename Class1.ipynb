```python
#drop 1st two character strings from crabs to make it usable by gmm..
#crabs = np.genfromtxt("crabs.csv", dtype=None, delimiter=',', names=True) 
crabs = pd.read_stata("crabs.dta")
#three ways to do the same thing (prep data)
crabs_mat = crabs.as_matrix()[:,3:].astype(np.float)
crabs_mat = crabs[['FL', 'RW', 'CL', 'CW', 'BD']].as_matrix()
crabs_mat = crabs[crabs.columns[3:]].as_matrix()
```

```python
def num_corr(x): 
    """
    permute cols of x to find maximal diagonal sum (maximal correctly classified)
    """
    import itertools as it
    max_t = 0
    n = x.shape[1]
    perm = range(n)
    for i in it.permutations(range(n)):
        new_t = np.diag(x[:,i]).sum()
        if new_t>max_t:
            max_t = new_t
            perm = i
    return max_t,perm,x[:,perm]
```

```python
g = mixture.GMM(covariance_type='full',n_components=4) #for some reason only full works
g.fit(crabs_mat) 
g.bic(crabs_mat)
#assign unique codes to crabs
sp=pd.get_dummies(crabs["species"])
sx=pd.get_dummies(crabs["sex"])
target=pd.concat([sp,sx],axis=1)
target["code"] = 2*target["Blue"]+target["Male"]
gps = g.predict(crabs_mat)
cm = confusion_matrix(target["code"],gps)
print cm
print 'latent means:\n',np.around(g.means_,1),'\n\nlatent covars:\n',np.around(g.covars_,1)
print '\nbic=',round(g.bic(crabs_mat),2)
mx,perm,arr=num_corr(cm)
print 'Maximum correct=',mx
print 'Confusion Matrix:\n',arr
plt.matshow(arr)
```
